### CAR DEKHO ANALYSIS

##### Name (car model), year (manufacturing year), selling price, kilometres driven, fuel type, seller type (individual or dealer), gearbox type (manual or automatic), and owner count (number of previous owners) are some of the key attributes that are included in the dataset located on the CARDEKHO website. The dataset also includes information about used cars that are currently available for purchase. This information makes it possible to do a study of what elements influence the selling price, such as the age of the vehicle, the mileage, the kind of fuel, and the type of seller. By analysing this data, it is possible to discover patterns and insights about the price of automobiles, the preferences of buyers, and the dynamics of the market
